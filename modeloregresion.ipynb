```python
import pymongo
from pprint import pprint
client = pymongo.MongoClient("mongodb+srv://tfm:tfm123@cluster0-9biaw.azure.mongodb.net/test?retryWrites=true&w=majority")
db = client.TFM
import datetime
import random
trafico=db.trafico
cursor = trafico.find()

valores=[]
encabezados=[]
for i in range(24):
    encabezados.append("hora "+str(i))
dias=['L','M','X','J','V','S','D']
for i in range(7):
    encabezados.append(dias[i])
encabezados.append("traficoactual")
encabezados.append("calle")
encabezados
#list(cursor)[0]
for c in cursor:
    valores.append([ c[encabezados[i]] for i in range(len(encabezados))])
valoresrdd=sc.parallelize(valores)
from pyspark.sql import Row
row_data = valoresrdd.map(lambda p: Row(
    hora_0=p[0],
    hora_1=p[1],
    hora_2=p[2],
    hora_3=p[3],
    hora_4=p[4],
    hora_5=p[5],
    hora_6=p[6],
    hora_7=p[7],
    hora_8=p[8],
    hora_9=p[9],
    hora_10=p[10],
    hora_11=p[11],
    hora_12=p[12],
    hora_13=p[13],
    hora_14=p[14],
    hora_15=p[15],
    hora_16=p[16],
    hora_17=p[17],
    hora_18=p[18],
    hora_19=p[19],
    hora_20=p[20],
    hora_21=p[21],
    hora_22=p[22],
    hora_23=p[23],
    L=p[24],
    M=p[25],
    X=p[26],
    J=p[27],
    V=p[28],
    S=p[29],
    D=p[30],
    traficoactual=p[31],
    calle=p[32]
))
valores_df = spark.createDataFrame(row_data)
valores_df.show()
valores_df.printSchema()
valores_df.describe().toPandas().transpose()
encabezados2=[]
for i in range(len(encabezados)):
    encabezados2.append(encabezados[i].replace(" ", "_"))
    print(encabezados[i].replace(" ", "_")+"=p["+str(i)+"],")
encabezados2.pop(32)
encabezados2.pop(31)
```

```python
from pyspark.ml.feature import VectorAssembler
# genera un vector a partir de los valores X
assembler = VectorAssembler(
        inputCols=encabezados2,
        outputCol='features')

datosprep =  assembler.transform(valores_df)
datosprep = datosprep.select(['traficoactual','features'])

print(datosprep)
```

```python
datosprep.take(2)
```

```python

[train,test]=datosprep.randomSplit([0.7,0.3])
from pyspark.ml.regression import LinearRegression
# solucion

lr = LinearRegression(maxIter=10, regParam=0.1, elasticNetParam=0.8, labelCol='traficoactual')
model_rl = lr.fit(train)
res_test = model_rl.transform(train)
res_test.show()
# Otro método: árboles de decisión
from pyspark.ml.regression import DecisionTreeRegressor

dt = DecisionTreeRegressor(labelCol='traficoactual')
dt_model = dt.fit(train)
dt_predictions = dt_model.transform(test)
from pyspark.ml.evaluation import RegressionEvaluator
metrics=["r2","mae","mse","rmse"]
evaluators = [(m,RegressionEvaluator(predictionCol="prediction", \
                 labelCol="traficoactual",metricName=m)) for m in metrics]
errors = [ (name, eval.evaluate(dt_predictions)) for name,eval in evaluators]
print(errors)
```
